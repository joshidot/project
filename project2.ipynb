```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits import mplot3d
```

```python
df = pd.read_csv('data/data.csv') 
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df['month'] = pd.to_datetime(df['date'], format='%y%M%d').dt.month
df
```

```python
df['hour'] = pd.to_datetime(df['time'], format='%H:%M').dt.hour
df
```

```python
n=145
```

```python
fig = plt.figure(figsize = [10,10])
ax = plt.axes(projection='3d')
ax.scatter3D(df['lati'], df['long'], df['magni'])
ax.set_xlabel('Latitude($^o$N)')
ax.set_ylabel('Longitude($^o$E)')
ax.set_zlabel('Magnitude')
plt.title('Earth Quake in 2015')
plt.show()
```

```python
plt.figure(figsize = [10,10])
plt.scatter(df['lati'], df['long'])
plt.xlabel('Latitude($^o$N)')
plt.ylabel('Longitude($^o$E)')
plt.title('Earth Quake in 2015')
plt.show()
```

```python
plt.figure(figsize = [10,8])
plt.scatter(df['month'],df['magni'])
plt.xlabel('Month')
plt.ylabel('Magnitude')
plt.xticks(rotation=45)
plt.title('Earth Quake in 2015')
```

```python
plt.figure(figsize = [10,8])
plt.scatter(df['hour'],df['magni'])
plt.xlabel('Hour')
plt.ylabel('Magnitude')
plt.xticks(rotation=45)
plt.title('Earth Quake in 2015')
```
